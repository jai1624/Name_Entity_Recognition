```python
#spacy english language
nlp_en = spacy.load("en_core_web_lg")
#spaCy pipeline for biomedical data
nlp_bio_data = spacy.load("en_core_sci_sm")
#biomedical corpus
nlp_lg = spacy.load('en_core_sci_lg')
#large biomedical corpus
nlp_md = spacy.load('en_core_sci_md')
#bert biomedical corpus
nlp_scibert = spacy.load('en_core_sci_scibert')
#Identify drug
med7=spacy.load("en_core_med7_lg")
# Model detect disease and chemical
nlp_bc = spacy.load("en_ner_bc5cdr_md")
#Model to detect cancer
nlp_bi = spacy.load("en_ner_bionlp13cg_md")
#Model to detect GGP,SO, TAXON
nlp_craft = spacy.load("en_ner_craft_md")
#Model to detect DNA,CELL,RNA
nlp_jnlp = spacy.load("en_ner_jnlpba_md")
```

```python
total_ner=med7.pipe_labels['ner'] + nlp_lg.pipe_labels['ner'] + nlp_bc.pipe_labels['ner'] + nlp_bi.pipe_labels['ner'] +nlp_en.pipe_labels['ner']+nlp_md.pipe_labels['ner']+nlp_scibert.pipe_labels['ner']
print(total_ner)
```

```python
def entity_color_generator(number_of_colors):
        color = ["#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)])
                 for i in range(number_of_colors)]
        return color
```

```python
total_number_entity=np.int(len(total_ner))
print(total_number_entity)
```

```python
genertate_color=entity_color_generator(total_number_entity)
print(genertate_color)
#prepare the options
col_dict = {}

for label, colour in zip(total_ner, genertate_color):  
    col_dict[label] = colour
options = {'ents': total_ner, 'colors':col_dict}
print(options)
```

```python
# Combine Entities and Labels
model_1_entity= med7(text_doc)
model_2_entity=nlp_lg (text_doc)
model_3_entity=nlp_bc(text_doc)
model_4_entity=nlp_bi(text_doc)
model_5_entity=nlp_en(text_doc)
model_6_entity=nlp_scibert(text_doc)
model_7_entity=nlp_md(text_doc)
################################################################Label
model_1_entity_label=[ent.label_ for ent in model_1_entity.ents]
model_2_entity_label=[ent.label_ for ent in model_2_entity.ents]
model_3_entity_label=[ent.label_ for ent in model_3_entity.ents]
model_4_entity_label=[ent.label_ for ent in model_4_entity.ents]
model_5_entity_label=[ent.label_ for ent in model_5_entity.ents]
model_6_entity_label=[ent.label_ for ent in model_6_entity.ents]
model_7_entity_label=[ent.label_ for ent in model_7_entity.ents]
######################################################## Text
model_1_entity_text=[ent.text for ent in model_1_entity.ents]
model_2_entity_text=[ent.text for ent in model_2_entity.ents]
model_3_entity_text=[ent.text for ent in model_3_entity.ents]
model_4_entity_text=[ent.text for ent in model_4_entity.ents]
model_5_entity_text=[ent.text for ent in model_5_entity.ents]
model_6_entity_text=[ent.text for ent in model_6_entity.ents]
model_7_entity_text=[ent.text for ent in model_7_entity.ents]
############################################################################
final_entity_text=model_1_entity_text + model_2_entity_text + model_3_entity_text + model_4_entity_text+model_5_entity_text
final_enity_label=model_1_entity_label + model_2_entity_label + model_3_entity_label + model_4_entity_label+model_5_entity_label
#final_entity_text=remove_duplicate_entry(final_entity_text)
print(final_entity_text)
#print(len(model_5_entity_text))
#print(len(model_5_entity_label))

#################
print([model_1_entity_text, model_1_entity_label ])
print([model_3_entity_text, model_3_entity_label ])
print([model_3_entity_text, model_3_entity_label ])
print([model_4_entity_text, model_4_entity_label ])
print([model_5_entity_text, model_5_entity_label ])
```

```python
intial_match_phrases=final_entity_text
print(intial_match_phrases)
intial_match_label=final_enity_label
print(intial_match_label)
matches = []
for labels, match_phrase in enumerate(intial_match_phrases):
    for item in re.finditer(match_phrase,text_doc):
        match = {}
        match['start'], match['end'] = item.span() 
        match['label'] = intial_match_label[labels] # The tag/label that you would like to display
        matches.append(match)
        #print(matches)
matches=sorted(matches, key=lambda e: (e['start'], -e['end']))
res_list = []
locs=[]
#print(max_value)
#print(max_index)
for i in range(len(matches)):
        curr_locs=matches[i]['start'],matches[i]['end']
        #maxPricedItem = max(curr_locs,key=lambda x: matches[i]['end'])
        #print(maxPricedItem)
        if curr_locs not in locs:
            locs.append(curr_locs)
            #print(locs)
            res_list.append(matches[i])
            #print(res_list)
        else:
            if res_list[-1]['label']=='ENTITY':
                #print(res_list)
                res_list[-1]=matches[i]
                #print(matches[i])
sentence = [{
    'text': text_doc,
    'ents': res_list
}]
print(res_list)
```
